```python
# Import matplotlib's pyplot for plotting and numpy for numerical operations
import matplotlib.pyplot as plt
import numpy as np

# Generate some data: x is an array of 100 points between 0 and 10, y is the sine of these points
x = np.linspace(0, 10, 100)  # np.linspace generates linearly spaced vectors, which is useful for creating domains for functions
y = np.sin(x)  # np.sin calculates the sine of each element in x

# Create a figure and an axes object, allowing for plotting
fig, ax = plt.subplots()  # plt.subplots() creates a figure and one subplot by default, returning figure and axes objects

# Plotting a line plot on the axes
ax.plot(x, y, label='Sine Wave', color='blue')  # ax.plot() plots y versus x as lines and/or markers. Here, 'label' names the data series

# Adding scatter plot on the same axes
ax.scatter(x, y*2, color='red', label='Sine Wave Doubled')  # ax.scatter() plots y versus x as dots. Here, y*2 doubles the y values for visual contrast

# Set titles and labels for the axes
ax.set_title('Line and Scatter Plot')  # ax.set_title() sets the title of the axes
ax.set_xlabel('X axis')  # ax.set_xlabel() sets the label for the x-axis
ax.set_ylabel('Y axis')  # ax.set_ylabel() sets the label for the y-axis

# Enable legend to help identify the plots
ax.legend()  # ax.legend() displays a legend that identifies each dataset using the labels defined in plot() and scatter()

# Show the plot on the screen
plt.show()  # plt.show() displays the figure, a necessary call to actually see the plot when not using interactive environments

```
