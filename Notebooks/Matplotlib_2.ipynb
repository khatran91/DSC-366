```python
# Import necessary libraries for plotting and data manipulation
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)  # Generate 100 linearly spaced numbers between 0 and 10
y = np.sin(x)                # Compute the sine of these numbers
z = np.random.normal(size=100)  # Generate 100 random numbers from a normal distribution
categories = ['Apple', 'Banana', 'Cherry', 'Date']  # Categories for bar chart

# Create a figure to host all plots
plt.figure(figsize=(14, 10))  # Initialize a new figure window with a specific size (width, height)

# Create a line plot
plt.subplot(2, 2, 1)  # Divide the figure into a 2x2 grid, and start plotting in the first cell
plt.plot(x, y, label='Sin(x)')  # Plot y vs. x as a line plot with a label
plt.title('Line Plot')  # Add a title to the plot
plt.xlabel('X axis')  # Label the x-axis
plt.ylabel('Y axis')  # Label the y-axis
plt.legend()  # Display a legend with plot labels
plt.grid(True)  # Enable a grid to improve readability

# Create a scatter plot
plt.subplot(2, 2, 2)  # Prepare to plot in the second cell of the 2x2 grid
plt.scatter(x, y, color='red', label='Sin(x) Points')  # Create a scatter plot of the same data
plt.title('Scatter Plot')  # Title for the scatter plot
plt.xlabel('X axis')  # Label the x-axis
plt.ylabel('Y axis')  # Label the y-axis
plt.legend()  # Show the legend

# Create a histogram
plt.subplot(2, 2, 3)  # Plot in the third cell
plt.hist(z, bins=20, color='green', label='Normal Distribution')  # Create a histogram with 20 bins
plt.title('Histogram')  # Set the title of the histogram
plt.xlabel('Value')  # Label the x-axis with 'Value'
plt.ylabel('Frequency')  # Label the y-axis with 'Frequency'
plt.legend()  # Display the legend

# Create a bar chart
plt.subplot(2, 2, 4)  # Plot in the fourth cell
bar_heights = np.random.randint(1, 10, size=4)  # Random heights for each bar
plt.bar(categories, bar_heights, color='blue', label='Fruit Counts')  # Create a bar chart
plt.title('Bar Chart')  # Title for the bar chart
plt.xlabel('Categories')  # Label x-axis with categories
plt.ylabel('Counts')  # Label y-axis with counts
plt.legend()  # Show legend

# Adjust layout and display the plots
plt.tight_layout()  # Adjust the layout to make sure there is no content overlap
plt.show()  # Display all the subplots in the figure
```
