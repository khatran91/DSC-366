```python
import matplotlib.pyplot as plt
weight = 0.3
input_score = 2
ground_truth_prediction = 0.4
alpha = 0.1

error = []
prediction = []

for iteration in range(20):
    neural_network_prediction = input_score * weight
    square_error = (neural_network_prediction - ground_truth_prediction ) ** 2
    error.append(square_error)
    prediction.append(neural_network_prediction)

    direction_and_amount = (neural_network_prediction - ground_truth_prediction ) * input_score
    weight = weight - (alpha * direction_and_amount)


    print("Error:" + str(square_error) + " Prediction: " + str(neural_network_prediction))

plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(error, marker='o')
plt.title('Error Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Error')

plt.subplot(1, 2, 2)
plt.plot(prediction, marker='o')
plt.title('Predictions Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Prediction')

plt.tight_layout()
plt.show()
```
