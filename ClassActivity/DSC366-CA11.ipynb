```python
# Import necessary libraries for plotting and data manipulation
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)  # Generate 100 linearly spaced numbers between 0 and 10
y = np.sin(x)                # Compute the sine of these numbers
y_cos = np.cos(x)            # Compute the cosine of these numbers
z = np.random.normal(size=100)  # Generate 100 random numbers from a normal distribution
categories = ['Apple', 'Banana', 'Cherry', 'Date']  # Categories for bar chart
bar_heights = np.random.randint(1, 15, size=4)  # Random heights for original bars
bar_heights2 = np.random.randint(1, 15, size=4)  # Random heights for comparison bars

# Create a figure to host all plots
plt.figure(figsize=(14, 10))  # Initialize a new figure window with a specific size (width, height)

# Line Plot with markers and two functions
plt.subplot(2, 2, 1)
plt.plot(x, y, '-b', label='sin(x)', marker='o')  # Sine plot with markers
plt.plot(x, y_cos, '--r', label='cos(x)')  # Cosine plot with dashed line
plt.title('Line Plot with Sine and Cosine')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Scatter plot with variable size and color
plt.subplot(2, 2, 2)
sizes = 50 * abs(y)  # Size proportional to the absolute value of y
plt.scatter(x, y, s=sizes, c=sizes, cmap='viridis', alpha=0.6,
            edgecolor='black', label='Sin(x) Magnitude')
plt.title('Scatter Plot with Variable Marker Size')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Histogram with different bins and annotation
plt.subplot(2, 2, 3)
plt.hist(z, bins=20, color='gray', label='Normal Distribution')
mean_z = np.mean(z)
plt.axvline(mean_z, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_z, 10, 'Mean: {:.2f}'.format(mean_z), rotation=0, color='red')
plt.title('Histogram of Z')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Bar chart comparison
plt.subplot(2, 2, 4)
index = np.arange(len(categories))  # the x locations for the groups
bar_width = 0.35  # the width of the bars
plt.bar(index, bar_heights, bar_width, label='Set 1', color='b')
plt.bar(index + bar_width, bar_heights2, bar_width, label='Set 2', color='r')
plt.title('Bar Chart Comparison')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.xticks(index + bar_width / 2, categories)  # Positioning category labels
plt.legend()

# Adjust layout and display the plots
plt.tight_layout()  # Adjust the layout to make sure there is no content overlap
plt.show()  # Display all the subplots in the figure
```
